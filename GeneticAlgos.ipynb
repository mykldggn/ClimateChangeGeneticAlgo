```python
from tpot import TPOTClassifier, TPOTRegressor
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import kagglehub as kh
import os
```

```python
!pip install kaggle
import os
os.environ['KAGGLE_CONFIG_DIR'] = "/content"
dataset_file = "GlobalLandTemperaturesByCity.csv"
if not os.path.exists(dataset_file):
    # Download the dataset if it doesn't exist
    !kaggle datasets download berkeleyearth/climate-change-earth-surface-temperature-data
    # Unzip the downloaded file
    !unzip climate-change-earth-surface-temperature-data.zip
# Load the CSV file into a DataFrame
df = pd.read_csv(dataset_file)
print(df.head())
print(df.info())
```

```python
df.head()
```

```python
#shuffle
df_shuffle = df.iloc[np.random.permutation(len(df))]
```

```python
data = df_shuffle.reset_index(drop=True)
```

```python
#Store 2 Classes
```

```python
# Create 'NearEquator' column (1 if near equator, 0 otherwise)
data['Latitude'] = pd.to_numeric(data['Latitude'], errors='coerce') # Convert 'Latitude' to numeric
data['NearEquator'] = (data['Latitude'].abs() < 15).astype(int)  # Changed to 15
# Now you can use the 'NearEquator' column for stratified splitting:
data_class = data['NearEquator'].values
training_indices, validation_indices = training_indices, testing_indices = train_test_split(
    data.index, stratify=data_class, train_size=0.75, test_size=0.25
)
```

```python
#Use Genetic Programming to find best ML Model/Hyperparameters
```

```python
tpot = TPOTClassifier(generations=5, population_size=50, verbosity=2, random_state=69)
tpot
```

```python
for column in data.drop('NearEquator', axis=1).columns:
    try:
        data[column] = pd.to_numeric(data[column], errors='coerce')
    except ValueError:
        # Handle columns that cannot be converted to numeric (e.g., dates or strings)
        print(f"Column '{column}' could not be converted to numeric. Consider encoding or dropping it.")
```

```python
tpot.fit(data.drop('NearEquator', axis=1).loc[training_indices].values,
         data.loc[training_indices, 'NearEquator'].values)
```

```python
tpot.score(data.drop('NearEquator', axis=1).loc[validation_indices].values,
          data.loc[validation_indices, 'NearEquator'].values)
```
